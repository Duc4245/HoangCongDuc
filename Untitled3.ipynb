```python
import pandas as pd

print("="*80)
print("BÀI 1: THAO TÁC VÀ XỬ LÝ DỮ LIỆU ĐIỂM THI ĐẠI HỌC")
print("="*80)

# 3. Đọc dữ liệu
df = pd.read_csv('/content/dulieuxettuyendaihoc.csv')
print(f"\nKích thước dữ liệu: {df.shape[0]} dòng x {df.shape[1]} cột\n")

print("10 DÒNG ĐẦU TIÊN:")
print(df.head(10))

print("\n10 DÒNG CUỐI CÙNG:")
print(df.tail(10))

# 1. Phân loại dữ liệu định tính và định lượng
print("\n" + "="*60)
print("1. PHÂN LOẠI DỮ LIỆU")
print("="*60)
dl_dinh_luong = ['T1','L1','H1','S1','V1','X1','D1','N1','T2','L2','H2','S2','V2','X2','D2','N2',
                 'T6','L6','H6','S6','V6','X6','D6','N6','DH1','DH2','DH3']
dl_dinh_tinh = ['GT','DT','KV','KT']

print("Dữ liệu ĐỊNH LƯỢNG:", dl_dinh_luong)
print("Dữ liệu ĐỊNH TÍNH  :", dl_dinh_tinh)

# 2. Thang đo cho từng biến
print("\n" + "="*60)
print("2. THANG ĐO CỦA TỪNG BIẾN")
print("="*60)
print("GT, DT, KV, KT : Nominal / Ordinal")
print("Tất cả các cột điểm (T*, L*, H*, S*, V*, X*, D*, N*, DH*) : Ratio")

# ====================== XỬ LÝ DỮ LIỆU THIẾU ======================

# 4. Xử lý cột DT (Dân tộc)
print("\n" + "="*60)
print("4. XỬ LÝ DỮ LIỆU THIẾU CỘT DT")
print("="*60)
print("Số missing DT:", df['DT'].isnull().sum())

print("\nBảng tần số:")
print(df['DT'].value_counts(dropna=False))
print("\nTần suất (%):")
print((df['DT'].value_counts(dropna=False, normalize=True)*100).round(2))

df['DT'] = df['DT'].fillna(0).astype(int)
print("→ Đã điền 0 vào missing DT")

# 5. Xử lý missing cho cột T1 bằng Mean
print("\n" + "="*60)
print("5. XỬ LÝ DỮ LIỆU THIẾU CỘT T1 (bằng Mean)")
print("="*60)
print("Số missing T1:", df['T1'].isnull().sum())
print("Mean của T1 =", df['T1'].mean())

df['T1'] = df['T1'].fillna(df['T1'].mean())
print("→ Đã thay missing T1 bằng Mean")

# 6. Xử lý missing cho tất cả các cột điểm còn lại bằng Mean
print("\n" + "="*60)
print("6. XỬ LÝ MISSING TẤT CẢ CÁC CỘT ĐIỂM CÒN LẠI")
print("="*60)

diem_cols = ['L1','H1','S1','V1','X1','D1','N1',
             'T2','L2','H2','S2','V2','X2','D2','N2',
             'T6','L6','H6','S6','V6','X6','D6','N6',
             'DH1','DH2','DH3']

for col in diem_cols:
    if df[col].isnull().sum() > 0:
        mean_val = df[col].mean()
        df[col] = df[col].fillna(mean_val)
        print(f"Đã fill missing {col:4} bằng mean = {mean_val:.2f}")

# 7. Tạo TBM1, TBM2, TBM3
df['TBM1'] = (df['T1']*2 + df['L1'] + df['H1'] + df['S1'] + df['V1']*2 + df['X1'] + df['D1'] + df['N1']) / 10
df['TBM2'] = (df['T2']*2 + df['L2'] + df['H2'] + df['S2'] + df['V2']*2 + df['X2'] + df['D2'] + df['N2']) / 10
df['TBM3'] = (df['T6']*2 + df['L6'] + df['H6'] + df['S6'] + df['V6']*2 + df['X6'] + df['D6'] + df['N6']) / 10

# 8. Tạo XL1, XL2, XL3
def xeploai(tbm):
    if tbm < 5.0: return 'Y'
    elif tbm < 6.5: return 'TB'
    elif tbm < 8.0: return 'K'
    elif tbm < 9.0: return 'G'
    else: return 'XS'

df['XL1'] = df['TBM1'].apply(xeploai)
df['XL2'] = df['TBM2'].apply(xeploai)
df['XL3'] = df['TBM3'].apply(xeploai)

# 9. US_TBM1, US_TBM2, US_TBM3 (thang 4.0)
df['US_TBM1'] = df['TBM1'] / 2.5
df['US_TBM2'] = df['TBM2'] / 2.5
df['US_TBM3'] = df['TBM3'] / 2.5

# 10. KQXT
def ketqua(row):
    kt = row['KT']
    if kt in ['A', 'A1']:
        diem = (row['DH1']*2 + row['DH2'] + row['DH3']) / 4
    elif kt == 'B':
        diem = (row['DH1'] + row['DH2']*2 + row['DH3']) / 4
    else:
        diem = (row['DH1'] + row['DH2'] + row['DH3']) / 3
    return 1 if diem >= 5.0 else 0

df['KQXT'] = df.apply(ketqua, axis=1)

# 11. Lưu file
df.to_csv('processed_dulieuxettuyendaihoc.csv', index=False, encoding='utf-8-sig')
print("\n" + "="*60)
print("HOÀN THÀNH! File đã được lưu: processed_dulieuxettuyendaihoc.csv")
print("="*60)

# Hiển thị kết quả một số cột quan trọng
print("\nKết quả mẫu:")
print(df[['TBM1','XL1','US_TBM1','KQXT']].head(10).round(3))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Cài đặt hiển thị
pd.set_option('display.max_columns', None)
plt.rcParams['figure.figsize'] = (12, 6)
sns.set(style="whitegrid")
plt.rcParams['font.sans-serif'] = ['Arial', 'DejaVu Sans']

# ====================== ĐỌC DỮ LIỆU ======================
df = pd.read_csv('processed_dulieuxettuyendaihoc.csv')
print("Đọc file thành công! Kích thước:", df.shape)

# Tạo biến phân loại T1 (dùng cho Phần 4)
def phanlop_t1(x):
    if x < 5.0: return 'Kém'
    elif x < 7.0: return 'Trung bình'
    elif x < 8.0: return 'Khá'
    else: return 'Giỏi'

df['PhanlopT1'] = df['T1'].apply(phanlop_t1)

print("\n=== BÀI TRỰC QUAN HÓA DỮ LIỆU ĐÃ HOÀN THÀNH ===")
# ======================Phần 1=======================
print("\n" + "="*70)
print("PHẦN 1: THỐNG KÊ DỮ LIỆU")
print("="*70)

print("1. DH1 sắp xếp tăng dần (10 giá trị đầu):")
print(df['DH1'].sort_values().head(10))

print("\n2. DH2 sắp xếp tăng dần theo GT:")
print(df.sort_values(by=['GT','DH2'])[['GT','DH2']].head(10))

print("\n3. Pivot Table DH1 theo KT:")
print(df.pivot_table(values='DH1', index='KT',
                     aggfunc=['count','sum','mean','median','min','max','std']))

print("\n4. Pivot Table DH1 theo KT & KV:")
print(df.pivot_table(values='DH1', index=['KT','KV'],
                     aggfunc=['count','mean','median','min','max','std']))
#=====================Phần 2, 3=======================
print("\n" + "="*70)
print("PHẦN 2 & 3: TRÌNH BÀY VÀ TRỰC QUAN HÓA")
print("="*70)

# GT
print("Bảng tần số GT:")
print(df['GT'].value_counts())

# Một số biểu đồ chính
sns.countplot(data=df, x='GT')
plt.title('Phân bố Giới tính')
plt.show()

# KQXT theo KT
sns.countplot(data=df, x='KT', hue='KQXT')
plt.title('Kết quả xét tuyển theo Khối thi')
plt.show()

# Số lượng theo KV
pd.crosstab(df['KV'], df['KT']).plot(kind='bar')
plt.title('Số lượng thí sinh theo Khu vực và Khối thi')
plt.show()

# Đậu rớt theo GT, KV, DT
for col in ['GT', 'KV', 'DT']:
    sns.countplot(data=df, x=col, hue='KQXT')
    plt.title(f'Đậu/Rớt theo {col}')
    plt.show()
#=====================Phần 4=======================
print("\n" + "="*70)
print("PHẦN 4: TRỰC QUAN HÓA NÂNG CAO")
print("="*70)

# 1. Histogram T1
plt.figure(figsize=(10,6))
plt.hist(df['T1'], bins=25, color='skyblue', edgecolor='black', alpha=0.85)
plt.title('Histogram phân bố điểm T1', fontsize=14)
plt.xlabel('T1', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.grid(True, alpha=0.3)
plt.show()

# 2. Multiple Line theo PhanlopT1
plt.figure(figsize=(12,7))
colors = {'Giỏi':'blue', 'Khá':'green', 'Trung bình':'purple', 'Kém':'red'}

for label, group in df.groupby('PhanlopT1'):
    sns.kdeplot(data=group['T1'], label=label, linewidth=2.5, color=colors[label])
plt.title('Multiple Line Plot - Phân bố T1 theo PhanlopT1', fontsize=14)
plt.xlabel('T1', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend(title='Phân loại T1')
plt.grid(True, alpha=0.3)
plt.show()

# 3. Drop-line / Strip plot
plt.figure(figsize=(12,8))
sns.stripplot(data=df, x='T1', y='PhanlopT1', hue='PhanlopT1',
              jitter=True, size=8, palette=colors, linewidth=1)
plt.title('Drop-line Plot - T1 theo PhanlopT1', fontsize=14)
plt.xlabel('T1', fontsize=12)
plt.ylabel('Phân loại T1', fontsize=12)
plt.grid(True, axis='x', alpha=0.3)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
